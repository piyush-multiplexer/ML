```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import ensemble
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cancer_reg.csv')
df.head()
```

```python
# missing value handling
df['PctSomeCol18_24'] = df['PctSomeCol18_24'].fillna(df['PctSomeCol18_24'].mean())
df['PctEmployed16_Over'] = df['PctEmployed16_Over'].fillna(df['PctEmployed16_Over'].mean())
df['PctPrivateCoverageAlone'] = df['PctPrivateCoverageAlone'].fillna(df['PctPrivateCoverageAlone'].mean())
```

```python
y = df['TARGET_deathRate']
y.head()
```

```python
x = df
x = x.drop('TARGET_deathRate', axis=1)
x = x.drop('binnedInc', axis=1) #removed binnedInc field
x.head()
x = pd.get_dummies(x)
x.head()
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y)
```

```python
lr = linear_model.LinearRegression()
lr.fit(train_x,train_y)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
print('new data score: ',lr.score(test_x,test_y), 'same data score: ',lr.score(train_x,train_y))
```
